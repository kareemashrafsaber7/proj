```python
import pandas as pd
import glob
import os
import logging
```

```python
logging.basicConfig(
    filename='project_log.txt',
    level=logging.INFO,
    format='%(asctime)s -%(message)s'
)
print ('Starting ETL Process...')
logging.info('ETL Started')

try:
    dataframes = []
    files = glob.glob(r'C:\Users\Admin\Desktop\Project\*')
    print (f'Files found successfully,{len(files)}')
    for file in files:
        if file.endswith('.csv'):
            print (f'Loading csv file,{file}')
            df = pd.read_csv(file)
            dataframes.append(df)
            logging.info('DataFrame for CSV file is created successfully!')
        elif file.endswith('.json'):
            print (f'Loading json file,{file}')
            df=pd.read_json(file)
            dataframes.append(df)
            logging.info('DataFrame for JSON file is created successfully!')
    if len(dataframes)>0:
        complete_df = pd.concat(dataframes, ignore_index=True)
        print ("Data Extracted Successfully!")
        logging.info('Data Extracted Successfully!')
    else:
        raise ValueError ('No valid [csv/json] files were found in directory!')
    print ('Data cleaning starting up...')
    complete_df["date"] = pd.to_datetime(complete_df["date"])
    complete_df['product']=complete_df['product'].str.title()
    if complete_df['price'].isnull().sum() > 0:
        rows_before = len(complete_df)
        complete_df.dropna(subset=['price'],inplace=True)
        rows_after = len(complete_df)
        print (f'Successfully dropped {rows_before-rows_after} null price values!')
        logging.warning(f'Dropped {rows_before-rows_after} null price values!')
    complete_df['Total Cost']=complete_df['quantity']*complete_df['price']
    complete_df.drop_duplicates(inplace=True)
    complete_df['quantity']=complete_df['quantity'].fillna(0)
    print ('Data transformation was completed successfully!')
    if not os.path.exists('C:\\Users\\Admin\\Desktop\\Project\\output_data'):
        os.makedirs(r'C:\Users\Admin\Desktop\Project\output_data')
    output_dir = "C:\\Users\\Admin\\Desktop\\Project\\output_data\\ready_data.csv"
    complete_df.to_csv(output_dir, index=False)
    print (f'Data was loaded successfully!✅ you can find it at{output_dir}.')
except Exception:
    print (f"Error: Something went wrong, {Exception}")
    logging.info(f"Error: Something went wrong, {Exception}")
```
